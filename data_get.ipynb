```python
import requests
from bs4 import BeautifulSoup
import re
import csv


url = "https://www.fmprc.gov.cn/web/wjdt_674879/fyrbt_674889/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

findli=re.compile(r'<a href="(.*?)"> target="_blank">',re.S)

# 找到所有li标签中的符合条件的a标签并提取链接
date_links = [url + link.get('href') for link in soup.find_all('li')
              for link in link.find_all('a', href=True, target="_blank")
              if len(link.get('href')) == len("./202409/t20240911_11489002.shtml")]

# for link in date_links:
#     print(link)
with open('links.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for link in date_links:
        writer.writerow([link])
        
```

```python
# ruguo更复杂点用封装
questionscsv = []
with open('links.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        link = row[0]
        response = requests.get(link)
        response.encoding = 'utf-8'
        soup = BeautifulSoup(response.text, 'html.parser')
        # 提取包含“记者”字样的问题
        questions = [tag.text for tag in soup.find_all('strong') if "记者" in tag.text]
        for question in questions:
            #这里是去掉记者，去不去未知
            #question = question.replace(question.split('记者')[0] + '记者：', '')
            questionscsv.append(question)
            print(question)
with open('ques.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for question in questionscsv:
        writer.writerow([question])
```
