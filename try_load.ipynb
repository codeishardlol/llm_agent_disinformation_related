```python
from langchain.chains import SimpleSequentialChain

first_prompt = ChatPromptTemplate.from_template(
    '''信息迷雾有以下一级范畴的某一种或者多种特征:
    信息迷雾源头（创建者、提问者）、目的（成本目的、公关目的）、手段（政策手段、舆论手段、军事手段、外交手段、其他手段）、内容（政治、社会、经济、科技）,括号内为二级范畴。
    请分别给 {ques} 这句话做出标注，标注文本表示一级范畴的哪一种'''
)
chain_one = LLMChain(llm=llm, prompt=first_prompt, verbose=False)

second_prompt = ChatPromptTemplate.from_template(
    "如果文本属于某种范畴，请标注它属于二级范畴的哪一种: {realm1}"
)
chain_two = LLMChain(llm=llm, prompt=second_prompt, verbose=False)

# 目前可以要求它给出解释，也可以让他不要解释属于哪一种
# 使用SimpleSequentialChain将两个Chain串联起来，其中每个Chain都只支持一个输入和一个输出，根据chains列表中的顺序，将前一个Chain的输出作为下一个Chain的输入
overall_simple_chain = SimpleSequentialChain(
    chains=[chain_one, chain_two],
    verbose=False
)

ques = "日本广播协会记者：关于日本男童在深圳遇袭身亡事，王毅外长在纽约会见日本外相上川阳子时表示，“日方应当冷静理性看待，避免政治化和扩大化”。发言人能否详细说明“政治化和扩大化”的含义？"
res = overall_simple_chain.run(ques)
print(res)
```

```python
from langchain.chains import SequentialChain

#没有引入信息迷雾的介绍，挖个坑先
prompt1 = ChatPromptTemplate.from_template(
    '''信息迷雾属于以下一级范畴的某一种或者多种范畴：
    信息迷雾的源头、信息迷雾的目的、信息迷雾的手段、信息迷雾的内容。
    请对该句子: {ques_name}判断是否属于上述四种范畴的某一种或者多种,不要给出分析只输出属不属于四个范畴'''
)
chain1 = LLMChain(
    llm=llm,
    prompt=prompt1,
    output_key="realm1",
    verbose=False
)
overall_chain1 = SequentialChain(
    chains=[chain1],
    input_variables=["ques_name"],
    output_variables=["realm1"],
    verbose=False
)

prompt2 = ChatPromptTemplate.from_template(
    '''以下为更细致的范畴，括号内为一级范畴的二级范畴：
    源头（创建者、提问者）。
    请根据上述的标准对{realm1}标注其二级范畴，
    不要给出分析，只输出属不属于源头范畴，如果属于标注其属于二级范畴的哪一种'''
)
chain2 = LLMChain(
    llm=llm,
    prompt=prompt2,
    output_key="realm11"
)

prompt3 = ChatPromptTemplate.from_template(
   '''以下为更细致的范畴，括号内为一级范畴的二级范畴：
    目的（成本目的、公关目的）。
    请根据上述的标准对{realm1}标注其二级范畴，
    不要给出分析，只输出属不属于目的范畴，如果属于标注其属于二级范畴的哪一种'''
)
chain3 = LLMChain(
    llm=llm,
    prompt=prompt3,
    output_key="realm22"
)

prompt4 = ChatPromptTemplate.from_template(
    '''以下为更细致的范畴，括号内为一级范畴的二级范畴：
   手段（政策手段、舆论手段、军事手段、外交手段、其他手段。
    请根据上述的标准对{realm1}标注其二级范畴，
    不要给出分析，只输出属不属于手段范畴，如果属于标注其属于二级范畴的哪一种'''
)
chain4 = LLMChain(
    llm=llm,
    prompt=prompt4,
    output_key="realm33"
)

prompt5 = ChatPromptTemplate.from_template(
    '''以下为更细致的范畴，括号内为一级范畴的二级范畴：
    内容（政治、社会、经济、科技）。
    请根据上述的标准对{realm1}标注其二级范畴，
    不要给出分析，只输出属不属于内容范畴，如果属于标注其属于二级范畴的哪一种'''
)
chain5 = LLMChain(
    llm=llm,
    prompt=prompt4,
    output_key="realm44"
)

overall_chain2 = SequentialChain(
    chains=[chain2, chain3, chain4,chain5],
    input_variables=["realm1"],
    output_variables=["realm11", "realm22",  "realm33", "realm44"],
    verbose=False
)
ques_name = "法新社记者：加拿大前外交官康明凯在接受加拿大广播公司采访时称，自2019年起他在中国被单独拘押近6个月，受到精神折磨。中方对此有何回应？"
res = overall_chain1(ques_name)
res=overall_chain2(res)
res
```

```python
#根据提示词的不同而选择不同的Chain进行执行，实现分支判断的作用。

from langchain.chains.llm import LLMChain
from langchain.prompts import PromptTemplate
from langchain.chains.router.llm_router import LLMRouterChain, RouterOutputParser
from langchain.chains.router.multi_prompt_prompt import MULTI_PROMPT_ROUTER_TEMPLATE as RounterTemplate

disinfor_template = """
你是一个经验丰富的虚假信息识别器，擅长解答关于甄别是否属于虚假信息的问题。
下面是需要你来解析的问题:
{input}
"""

disinforsen_template = """
你是一位语言结构内容大师，擅长解答关于句子成分的问题。
下面是需要你来回答的问题:
{input}
"""


prompt_infos = [
    {
        "key": "disinfor",
        "description": "适合回答关于信息迷雾甄别的问题",
        "template": disinfor_template,
    },
    {
        "key": "disinforration",
        "description": "适合回答关于句子成分剖析的问题",
        "template": disinforsen_template,
    }
]

#构建目标链chain_map（json格式），以提示信息prompt_infos中的key为key，以Chain为value
chain_map = {}

for info in prompt_infos:
    prompt = PromptTemplate(
        template=info['template'],
        input_variables=["input"]
    )
    print("目标提示:\n", prompt)
    
    chain = LLMChain(
        llm=llm,
        prompt=prompt,
        verbose=0
    )
    chain_map[info["key"]] = chain

#构建路由链router_chain
destinations = [f"{p['key']}: {p['description']}" for p in prompt_infos]
router_template = RounterTemplate.format(destinations="\n".join(destinations))
print("路由模板:\n", router_template)

router_prompt = PromptTemplate(
    template=router_template,
    input_variables=["input"],
    output_parser=RouterOutputParser(),
)
print("路由提示:\n", router_prompt)

router_chain = LLMRouterChain.from_llm(
    llm,
    router_prompt,
    verbose=True
)

## 【Step5】构建默认链 default_chain 
from langchain.chains import ConversationChain
default_chain = ConversationChain(
    llm=llm,
    output_key="text",
    verbose=True
)

## 【Step6】构建多提示链 MultiPromptChain
from langchain.chains.router import MultiPromptChain

chain = MultiPromptChain(
    router_chain=router_chain,
    destination_chains=chain_map,
    default_chain=default_chain,
    verbose=True
)

# 测试1
print(chain.run("乌通社记者：乌克兰总统泽连斯基在接受巴西媒体采访时称，中国、巴西关于政治解决乌克兰危机的“六点共识”具有“破坏性”，中国和巴西未与乌克兰讨论，呼吁乌方与占领乌领土的敌人达成妥协不可接受。发言人对此有何评论？"
))

```
